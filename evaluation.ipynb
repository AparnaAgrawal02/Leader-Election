```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

# Evaluating the HS,LCR and franklin for ring Network

```python
LCR_messages = []
HS_messages= []
LCR_messages.append(1)
Franklin_messages = []
Franklin_messages.append(5)
Franklin_messages.append(10)
# Read the LCR messages
with open('LCR_messages', 'r') as f:
    for line in f:
        LCR_messages.append(int(line.strip()))

# Read the HS messages
with open('HS_messages', 'r') as f:
    for line in f:
        HS_messages.append(int(line.strip()))

with open('out_franklin.txt', 'r') as f:
    for line in f:
        Franklin_messages.append(int(line.strip()))
# Plot the LCR messages
plt.plot(LCR_messages, label='LCR')

# Plot the HS messages
plt.plot(HS_messages, label='HS')

# Plot the Franklin messages
plt.plot(Franklin_messages, label='Franklin')


# Set the labels
plt.xlabel('Number of nodes')
plt.ylabel('Messages')

# Set the title
plt.title('Messages vs Number of nodes')

# Set the legend
plt.legend()

# Show the plot
plt.show()



```

# Complete Network

```python
messages_sent = []
messages_received = []

with open('out_bully.txt','r') as f:
    for line in f:
        data = line.split()
        messages_sent.append(int(data[0]))
        messages_received.append(int(data[1]))

plt.plot(messages_sent, label='Messages sent')
plt.plot(messages_received, label='Messages received')

plt.xlabel('Number of nodes')
plt.ylabel('Messages')

plt.title('Messages vs Number of nodes')

```
